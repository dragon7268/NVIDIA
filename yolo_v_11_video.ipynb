```python
# yt-dlp 설치
!pip install yt-dlp

import yt_dlp

# 유튜브 링크 입력
video_url = "https://www.youtube.com/watch?v=Fs_SKa4u4_8"

# yt-dlp 설정 (최고 화질로 다운로드)
ydl_opts = {
    'format': 'best',
    'outtmpl': 'downloaded_video.mp4'  # 저장될 파일 이름
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

video_path = "downloaded_video.mp4"
print(f"다운로드 완료: {video_path}")
```

```python
# YOLO 라이브러리 설치
!pip install ultralytics

# YOLO 모델 불러오기
from ultralytics import YOLO
import cv2

# YOLO v11 모델 로드 (가벼운 'n' 모델 사용)
model = YOLO('yolo11n.pt')

# 동영상 열기
cap = cv2.VideoCapture("downloaded_video.mp4")

# 출력 동영상 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, cap.get(cv2.CAP_PROP_FPS),
                      (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

# 프레임별 객체 탐지 실행
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # YOLO 모델로 객체 탐지 수행
    results = model(frame)

    # 결과를 프레임에 적용 (바운딩 박스, 라벨 추가)
    annotated_frame = results[0].plot()

    # 결과 프레임 저장
    out.write(annotated_frame)

cap.release()
out.release()

print("📌 객체 탐지 완료! 결과 영상: output_video.mp4")
```
