```python
# yt-dlp ÏÑ§Ïπò
!pip install yt-dlp

import yt_dlp

# Ïú†ÌäúÎ∏å ÎßÅÌÅ¨ ÏûÖÎ†•
video_url = "https://www.youtube.com/watch?v=Fs_SKa4u4_8"

# yt-dlp ÏÑ§Ï†ï (ÏµúÍ≥† ÌôîÏßàÎ°ú Îã§Ïö¥Î°úÎìú)
ydl_opts = {
    'format': 'best',
    'outtmpl': 'downloaded_video.mp4'  # Ï†ÄÏû•Îê† ÌååÏùº Ïù¥Î¶Ñ
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

video_path = "downloaded_video.mp4"
print(f"Îã§Ïö¥Î°úÎìú ÏôÑÎ£å: {video_path}")
```

```python
# YOLO ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
!pip install ultralytics

# YOLO Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞
from ultralytics import YOLO
import cv2

# YOLO v11 Î™®Îç∏ Î°úÎìú (Í∞ÄÎ≤ºÏö¥ 'n' Î™®Îç∏ ÏÇ¨Ïö©)
model = YOLO('yolo11n.pt')

# ÎèôÏòÅÏÉÅ Ïó¥Í∏∞
cap = cv2.VideoCapture("downloaded_video.mp4")

# Ï∂úÎ†• ÎèôÏòÅÏÉÅ ÏÑ§Ï†ï
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, cap.get(cv2.CAP_PROP_FPS),
                      (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

# ÌîÑÎ†àÏûÑÎ≥Ñ Í∞ùÏ≤¥ ÌÉêÏßÄ Ïã§Ìñâ
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # YOLO Î™®Îç∏Î°ú Í∞ùÏ≤¥ ÌÉêÏßÄ ÏàòÌñâ
    results = model(frame)

    # Í≤∞Í≥ºÎ•º ÌîÑÎ†àÏûÑÏóê Ï†ÅÏö© (Î∞îÏö¥Îî© Î∞ïÏä§, ÎùºÎ≤® Ï∂îÍ∞Ä)
    annotated_frame = results[0].plot()

    # Í≤∞Í≥º ÌîÑÎ†àÏûÑ Ï†ÄÏû•
    out.write(annotated_frame)

cap.release()
out.release()

print("üìå Í∞ùÏ≤¥ ÌÉêÏßÄ ÏôÑÎ£å! Í≤∞Í≥º ÏòÅÏÉÅ: output_video.mp4")
```
