```python
import cv2
from ultralytics import YOLO

# 비디오 파일 경로 설정
video_path = '/content/video.mp4'  # 다운로드한 비디오 경로

# YOLOv8 모델 로드
model = YOLO('yolov8n.pt')  # yolov8n.pt, yolov8s.pt, yolov8m.pt 등 다양한 모델

# OpenCV로 비디오 캡처
cap = cv2.VideoCapture(video_path)

# 비디오가 열리면 계속해서 프레임을 처리
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break  # 비디오 끝에 도달하면 종료

    # YOLO 모델을 사용하여 객체 탐지
    results = model(frame)

    # 'results'는 리스트이므로 각 Result 객체를 꺼내서 plot() 호출
    for result in results:
        frame = result.plot()  # 객체 탐지된 결과를 이미지에 오버레이


    # 'q'를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
# 결과를 저장할 비디오 파일 설정
output_video_path = '/content/output_video.mp4'

# 비디오 파일로 저장을 위한 VideoWriter 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

# 비디오 처리 및 저장
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break  # 비디오 끝에 도달하면 종료

    # YOLO 모델을 사용하여 객체 탐지
    results = model(frame)

    # 'results'는 리스트이므로 각 Result 객체를 꺼내서 plot() 호출
    for result in results:
        frame = result.plot()  # 객체 탐지된 결과를 이미지에 오버레이

    # 탐지된 프레임을 비디오 파일로 저장
    out.write(frame)

# 비디오 캡처 및 저장 객체 해제
cap.release()
out.release()

print(f"Processed video saved to {output_video_path}")
```
